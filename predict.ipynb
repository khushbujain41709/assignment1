```python
import pandas as pd
import numpy as np
weather = pd.read_csv("Raipur.csv")
```

```python
weather
```

```python
# Droping irrelevant columns - according to the sensors we are using.
df_rnn = weather.copy()
df_rnn.drop(columns=['name', 'stations', 'dew', 'precipprob', 'preciptype', 'snow', 'snowdepth' , 'precip', 'cloudcover', 'visibility' , 'solarradiation', 	'solarenergy', 'uvindex', 'severerisk', 'windgust', 'icon'], inplace=True)
```

```python
# Auto-detect ISO 8601 and mixed formats
df_rnn['datetime'] = pd.to_datetime(df_rnn['datetime'], errors='coerce', utc=False)

# Sort by datetime
df_rnn.sort_values('datetime', inplace=True)

# Save cleaned dataset
df_rnn.to_csv("Raipur_cleaned.csv", index=False)
```
